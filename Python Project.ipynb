```python
#Checking for duplicates
df.duplicated()

# dropping unnecessary column
df.drop(columns = "记录数", inplace = True)
df

```

```python
#5.	What is the total sales value for the entire dataset?
df["Sales"].sum()
```

```python
# 6.	Which category has the highest total sales?
Sales_Category = df.groupby("Category")["Sales"].sum()
print(Sales_Category)
Sales_Category.sort_values(ascending=False)
```

```python
# 7.	Which sub-category has the lowest total sales?
Sales_SubCategory = df.groupby("Sub Category")["Sales"].sum().sort_values(ascending = True)
print(Sales_SubCategory)
```

```python
# 8.	What is the average sales value per order?
Avg_Order_Sales = df.groupby("Order ID")["Sales"].mean()
print(Avg_Order_Sales)
```

```python
# 9.	How do total sales vary across regions?
Sales_by_Regions = df.groupby("Region")["Sales"].sum()
print(Sales_by_Regions)
```

```python
import matplotlib.pyplot as plt
```

```python
# Sales by Region Visual
Sales_by_Regions.plot(kind = 'bar' , figsize = (7,5),  color = ["green","red", "yellow", "brown", "purple", "black", "blue"])
plt.show()
```

```python
# 10.	What is the total profit made?
Total_Profit = round(df["Profit"].sum(),2)
print (f"The total profit made is : {Total_Profit:.3f}")
```

```python
#11.	Which region generates the highest profit?
Profit_by_Region = round(df.groupby("Region")["Profit"].sum().sort_values(ascending = False),1)
print(Profit_by_Region)
```

```python
14.	# What is the total quantity of items sold?
Quantity_Sold = df["Quantity"].sum()
print(f"The total quantity of item sold is: {Quantity_Sold:,.2f}")
```

```python
# 15.	Which category sells the highest quantity of products?
Quantity_by_Category = df.groupby("Category")["Quantity"].sum()
Highest_category = Quantity_by_Category.idxmax()
Highest_Quantity = Quantity_by_Category.max()
print(f"Highest selling category: {Highest_category}")
print(f"Quantity sold: {Highest_Quantity:,}")
```

```python
#  16.	What is the average discount given across all orders?
Average_discout = df.groupby("Order ID")["Discount"].mean()
print(Average_discout)
```

```python
# 17.	Which region gives the highest average discount?
AverageDiscount_by_Region = df.groupby("Region")["Discount"].mean().sort_values(ascending = False)
print(AverageDiscount_by_Region)
```

```python
AverageDiscount_by_Region.plot(kind = "bar", figsize = (8,5), color = ["green","red", "yellow", "brown", "purple", "black", "blue", "maroon"])
    
plt.show()
```

```python
# converting year to integers
df["Year"] = df["Year"].astype(int)
```

```python
#18.	How does total sales change over time?
Sales_Over_Time = df.groupby("Year")["Sales"].sum()
print(Sales_Over_Time)
```

```python
df["Year"] = pd.to_datetime(df["Year"])
Sales_Over_Time.plot(figsize =(10,5))
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.title("Total Sales by Year")
plt.show()
```

```python
# 19.	Which month records the highest sales?
# df["Year"] = pd.to_datetime(df["Year"])
df["Month_Name"] = df["Year"].dt.month_name()
Monthly_sales = df.groupby("Month_Name")["Sales"].sum().sort_values(ascending = False)
print(Monthly_sales)
```

```python
# 20.	Is there a visible trend in sales growth or decline over time?
df["Year"] = df["Year"].astype(int)
df["Year"] = pd.to_datetime(df["Year"])
Sales_Over_Time.plot(figsize =(10,5))
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.title("Total Sales by Year")
plt.show()
```
